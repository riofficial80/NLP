```python
import pandas as pd
```

```python
# Sample corpus
corpus = ["cat", "dog", "fish", "cat", "dog"]
```

```python
# Create a vocabulary
vocab = sorted(set(corpus))  # e.g., ['cat', 'dog', 'fish']
print(vocab)
vocab_dict = {word: idx for idx, word in enumerate(vocab)}
print(vocab_dict)
```

```python
# One-hot encoding function
def one_hot_encode(word, vocab_dict):
    vec = [0] * len(vocab_dict)
    print(vec)
    vec[vocab_dict[word]] = 1
    return vec
```

```python
# Apply one-hot encoding
one_hot_vectors = [one_hot_encode(word, vocab_dict) for word in corpus]
```

```python
# Create DataFrame for visualization
one_hot_df = pd.DataFrame(one_hot_vectors, columns=vocab)
print("One-Hot Encoding:\n", one_hot_df)
```
